# **Sigmoid Function**

sigmoid-function.png

```python
import numpy as np

def sigmoid(x):
  return 1/(1+ np.exp(-x))

input = [15, 4, -14]

for x in input:
  print('{} --> {}'.format(x, sigmoid(x)))
```

# **Tanh Function**

tanh.png

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

input = [15, 4, -14]

for x in input:
  print('{} --> {}'.format(x, tanh(x)))
```

# **ReLU Function**

relu.png

```python
def relu(x):
  return max(0,x)

input = [15, 4, -14]

for x in input:
  print('{} --> {}'.format(x, relu(x)))
```

# **Leaky ReLU Function**

leakyrelu.png

```python
def leakyrelu(x):
  return max(0.1*x,x)

input = [15, 4, -14]

for x in input:
  print('{} --> {}'.format(x, leakyrelu(x)))
```

# **Softmax Function**

```python
def softmax(x):
  return np.exp(x) / np.exp(x).sum()

inputs = [1, 4, -2]

print(softmax(inputs))
```
